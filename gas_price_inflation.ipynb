```python
import os
import requests
import json
import pandas as pd
import prettytable
from pathlib import Path
import csv
```

```python
#API Call for 4 US Regions

headers = {'Content-type': 'application/json'}
series_id = ['APU010074714','APU020074714', 'APU030074714', 'APU040074714']
data = json.dumps({"seriesid": series_id,"startyear":"2015", "endyear":"2021"})
p = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data, headers=headers)
json_data = json.loads(p.text)
df = pd.json_normalize(json_data['Results']['series'], "data", ["seriesID"])
df
```

```python
#Setting the index of the df to seriesID
df.set_index("seriesID", inplace=True)

df
```

```python
#Reading the series_id.csv file with furhter information abour the series.
seriesid_df = pd.read_csv(
    Path("Resources/series_id.csv")
)
seriesid_df.head()
```

```python
seriesid_df.set_index("seriesID", inplace=True)

seriesid_df
```

```python
gas_df = pd.merge(df, seriesid_df, on="seriesID")
gas_df["area_code"].astype(int)
gas_df
```

```python
#Defining variables for commuting assumptions
#Assumming different commute miles roundtrip per work day per region
weekly_commute_west_before=150
weekly_commute_midwest_before=120
weekly_commute_northeast_before=75
weekly_commute_south_before=60
#Hybrid work model requires 2 days work in office compared to 5 days before
#Creating a variable to calculate delta
weekly_commute_change=1-(2/5)


```

```python
#Building a dataframe with commute data per area (0400=West, 0300=South ,0100=NorthEast, 0200=Midwest)
d = {'area_code': [100,200,300,400], 'Average Commute Before': [75,120,60,150]}
commute_df = pd.DataFrame(data=d,)

commute_df
```

```python
#Add Commute Savings
commute_df['Commute Savings']=commute_df['Average Commute Before']*weekly_commute_change
commute_df
```

```python
gas_df['area_code']=gas_df['area_code'].astype('int64')
gas_df['value']=gas_df['value'].astype('float')
```

```python
#Merging commute df with the gas prices df
final_df=pd.merge(gas_df,commute_df, how='left', on='area_code')
final_df
```

```python
#Dropping unnecessary columns
final_df.drop(columns=['latest','footnotes','begin_year','begin_period','end_year','end_period', 'footnote_codes','latest','period'])
final_df
```

```python
final_df['year'] = final_df['year'].astype('int64')
final_df = final_df[final_df['year'] > 2018]
```
